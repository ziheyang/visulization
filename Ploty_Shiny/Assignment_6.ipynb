```python
iso = px.data.gapminder()
iso2007 = iso[iso['year'] == 2007][['country', 'iso_alpha', 'pop']]
```

```python
labels = ['Happiness.Score', 'GDP', 'Social.Support', 'Health',
       'Freedom', 'Generosity', 'Corruption', 'Positive.Affect',
       'Negative.Affect', 'Confidence.in.Government', 'Democratic.Quality',
       'GINI.Index']
df[labels] = round(df[labels], 2)

iso = px.data.gapminder()
df_iso = pd.merge(df, iso2007, left_on=['Country'], right_on=["country"])
df_iso = df_iso.sort_values("Year")

import plotly.express as px
fig = px.scatter_geo(df_iso, locations="iso_alpha", 
                     hover_name="Country", 
                     animation_frame="Year",
                     projection="natural earth",
                     color='Happiness.Score',
                     color_continuous_scale='RdBu',
                     size="pop",
                     color_continuous_midpoint= np.average(df_iso['Happiness.Score']),
                     title = "Figure 1: Happiness Score in the World") 
fig.show()
fig.write_html('f1.html')
```

```python
# sunburst charts 
import plotly.express as px
import numpy as np

fig = 
(df2018, path=['Continent', 'Country'], values='pop',
                  color='Happiness.Score',
                  color_continuous_scale='RdBu',
                  color_continuous_midpoint= np.round(np.average(np.round(df2018['Happiness.Score'],2), weights=df2018['GDP']),2),
                 title="Figure 3: Happiness Score for Each Continent in Year 2018")
fig.show()
fig.write_html('f3.html')
```

```python
continents = ['Africa', 'Asia', 'Australia', 'Europe', 'North America', 'South America']

# negative: 'Generosity', 'Democratic.Quality',
lables = ['Happiness.Score', 'GDP', 'Social.Support', 'Health',
       'Freedom', 'Corruption', 'Positive.Affect',
       'Negative.Affect', 'Confidence.in.Government', 
       'GINI.Index']

gp_continent = df2018[lables + ['Continent']].groupby(['Continent']).mean()

gp_continent_tr = gp_continent.transpose()
# from sklearn.preprocessing import Normalizer
# gp_continent_tr = Normalizer(norm='l1').fit_transform(gp_continent_tr)

gp_continent_tr['Total'] = gp_continent_tr[continents].sum(axis=1)

new_continents = []
for c in continents:
    new_c = c + "_pc" #percentage
    new_continents.append(new_c)
    gp_continent_tr[new_c]  = round(gp_continent_tr[c]  * 100 / gp_continent_tr["Total"],2)
    
gp_continent_tr[new_continents]
```

```python
# Horizontal Bar Charts

import plotly.graph_objects as go

top_labels = ['Africa', 'Asia', 'Australia', 'Europe', 'North America', 'South America']

colors = ['rgba(38, 24, 74, 0.8)', 'rgba(71, 58, 131, 0.8)',
          'rgba(122, 120, 168, 0.8)', 'rgba(164, 163, 204, 0.85)',
          'rgba(190, 192, 213, 1)', 'rgba(38, 24, 74, 0.8)']

x_data = gp_continent_tr[new_continents].values

y_data = gp_continent_tr[new_continents].index

fig = go.Figure()

for i in range(0, len(x_data[0])):
    for xd, yd in zip(x_data, y_data):
        fig.add_trace(go.Bar(
            x=[xd[i]], y=[yd],
            orientation='h',
            marker=dict(
                color=colors[i],
                line=dict(color='rgb(248, 248, 249)', width=1)
            )
        ))

fig.update_layout(
    title = 'Figure 4: Matrics of Each Continent in Year 2018',
    xaxis=dict(
        showgrid=False,
        showline=False,
        showticklabels=False,
        zeroline=False,
        domain=[0.15, 1]
    ),
    yaxis=dict(
        showgrid=False,
        showline=False,
        showticklabels=False,
        zeroline=False,
    ),
    barmode='stack',
    paper_bgcolor='rgb(248, 248, 255)',
    plot_bgcolor='rgb(248, 248, 255)',
    margin=dict(l=120, r=10, t=140, b=80),
    showlegend=False,
)

annotations = []

for yd, xd in zip(y_data, x_data):
    # labeling the y-axis
    annotations.append(dict(xref='paper', yref='y',
                            x=0.14, y=yd,
                            xanchor='right',
                            text=str(yd),
                            font=dict(family='Arial', size=14,
                                      color='rgb(67, 67, 67)'),
                            showarrow=False, align='right'))
    # labeling the first percentage of each bar (x_axis)
    annotations.append(dict(xref='x', yref='y',
                            x=xd[0] / 2, y=yd,
                            text=str(xd[0])[:2] + '%',
                            font=dict(family='Arial', size=14,
                                      color='rgb(248, 248, 255)'),
                            showarrow=False))
    # labeling the first Likert scale (on the top)
    if yd == y_data[-1]:
        annotations.append(dict(xref='x', yref='paper',
                                x=xd[0] / 2, y=1.1,
                                text=top_labels[0],
                                font=dict(family='Arial', size=14,
                                          color='rgb(67, 67, 67)'),
                                showarrow=False))
    space = xd[0]
    for i in range(1, len(xd)):
            # labeling the rest of percentages for each bar (x_axis)
            annotations.append(dict(xref='x', yref='y',
                                    x=space + (xd[i]/2), y=yd,
                                    text=str(xd[i])[:2] + '%',
                                    font=dict(family='Arial', size=14,
                                              color='rgb(248, 248, 255)'),
                                    showarrow=False))
            # labeling the Likert scale
            if yd == y_data[-1]:
                annotations.append(dict(xref='x', yref='paper',
                                        x=space + (xd[i]/2), y=1.1,
                                        text=top_labels[i],
                                        font=dict(family='Arial', size=14,
                                                  color='rgb(67, 67, 67)'),
                                        showarrow=False))
            space += xd[i]

fig.update_layout(annotations=annotations)

fig.show()
fig.write_html('f4.html')
```

```python
# import graph objects as "go" and import tools
import plotly.graph_objs as go
from plotly import tools
import matplotlib.pyplot as plt

df2018top = df2018.head(10)
y_saving = [each for each in df2018top['GDP']]
y_net_worth  = [float(each) for each in df2018top['Happiness.Score']]
x_saving = [each for each in df2018top.Country]
x_net_worth  = [each for each in df2018top.Country]
trace0 = go.Bar(
                x=y_saving,
                y=x_saving,
                marker=dict(color='rgba(171, 50, 96, 0.6)',line=dict(color='rgba(171, 50, 96, 1.0)',width=1)),
                name='GDP',
                orientation='h',
)
trace1 = go.Scatter(
                x=y_net_worth,
                y=x_net_worth,
                mode='lines+markers',
                line=dict(color='rgb(63, 72, 204)'),
                name='Happiness.Score',
)
layout = dict(
                title='Figure 9: GDP & Happiness Score',
                yaxis=dict(showticklabels=True,domain=[0, 0.85]),
                yaxis2=dict(showline=True,showticklabels=False,linecolor='rgba(102, 102, 102, 0.8)',linewidth=2,domain=[0, 0.85]),
                xaxis=dict(zeroline=False,showline=False,showticklabels=True,showgrid=True,domain=[0, 0.42]),
                xaxis2=dict(zeroline=False,showline=False,showticklabels=True,showgrid=True,domain=[0.47, 1],side='top',dtick=25),
                legend=dict(x=0.029,y=1.038,font=dict(size=10) ),
                margin=dict(l=200, r=20,t=70,b=70),
                paper_bgcolor='rgb(248, 248, 255)',
                plot_bgcolor='rgb(248, 248, 255)',
)
annotations = []
y_s = np.round(y_saving, decimals=2)
# y_nw = np.rint(y_net_worth)
y_nw = np.round(y_net_worth, decimals=2)
# Adding labels
for ydn, yd, xd in zip(y_nw, y_s, x_saving):
    # labeling the scatter savings
    annotations.append(dict(xref='x2', yref='y2', y=xd, x=ydn - 4,text='{:,}'.format(ydn),font=dict(family='Arial', size=12,color='rgb(63, 72, 204)'),showarrow=False))
    # labeling the bar net worth
    annotations.append(dict(xref='x1', yref='y1', y=xd, x=yd + 3,text=str(yd),font=dict(family='Arial', size=12,color='rgb(171, 50, 96)'),showarrow=False))

layout['annotations'] = annotations

# Creating two subplots
fig = tools.make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_xaxes=True,
                          shared_yaxes=False, vertical_spacing=0.001)

fig.append_trace(trace0, 1, 1)
fig.append_trace(trace1, 1, 2)

fig['layout'].update(layout)

fig.show()
fig.write_html('f9.html')
```
